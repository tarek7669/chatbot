```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

#load model and tokenizer
model_name = "microsoft/phi-2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype = torch.bfloat16, device_map="auto")

#function to generate responses
def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")
    outputs = model.generate(**inputs, max_length=200)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

```python
print(generate_response("Hello, how are you?"))
```

```python
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from transformers import pipeline

#create text generation pipeline
hf_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=100)
llm = HuggingFacePipeline(pipeline=hf_pipeline)

#define prompt template
prompt = PromptTemplate(
    input_variables = ["question"],
    template = "You are a helpful chatbot. You act like a clone of Tarek Ashraf, your name is his name and his information is yours. answer these questions as best you can: {question}"
)

#create an LLMChain
llm_chain = LLMChain(prompt=prompt, llm=llm)

#function to chat using langchain
def chat_with_langchain(user_input):
    return llm_chain.run(user_input)
```

```python
#test the langchain function
print(chat_with_langchain("Tell me a fun fact about AI."))
```

```python
from langchain.text_splitter import CharacterTextSplitter
from langchain.document_loaders import PyPDFLoader

# Load and split text
loader = PyPDFLoader("/content/drive/MyDrive/Chatbot/data/data train.pdf")
documents = loader.load()

text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

#load embeddings
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

#store embeddings into chromaDB

vectorstore = Chroma.from_documents(docs, embedding_model)
```

```python
from langchain.chains import RetrievalQA

# create a retriever
retriever = vectorstore.as_retriever()
rag_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)

#chat function with RAG
def chat_with_rag(user_input):
    return rag_chain.run(user_input)
```

```python
#test chat with rag
print(chat_with_rag("what is AI?"))
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        break
    response = chat_with_rag(user_input)
    print("Chatbot:", response)
```
